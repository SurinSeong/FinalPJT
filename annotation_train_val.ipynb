## train, val 직접 나누기

```python
annotation_train = pd.read_csv(base_folder + '수린님/ResNet/annotation_train.csv')
annotation_val = pd.read_csv(base_folder + '수린님/ResNet/annotation_val.csv')

annotation = pd.concat([annotation_train, annotation_val])
annotation.head(2)
```

```python
# 턱선 처짐
chin_sagging = annotation[['ID', '성별', '나이', 'chin_sagging']].reset_index(drop=True)
chin_sagging.head(2)
```

```python
chin_sagging.describe()
```

```python
## 논문 내용을 통해 턱선 처짐이 성별과는 차이가 별로 없지만 나이와는 차이가 있다는 것을 알게 됨.
# 10, 20, 30, 40, 50, 60
def age_group(age):
    if age < 20:
        return 10
    elif age < 30:
        return 20
    elif age < 40:
        return 30
    elif age < 50:
        return 40
    elif age < 60:
        return 50
    else:
        return 60

chin_sagging['age_group'] = chin_sagging['나이'].apply(age_group)
chin_sagging.head(2)
```

```python
chin_sagging.groupby('chin_sagging').count()
```

```python
# 나이대별 턱선 처짐 등급 분포
print(chin_sagging.groupby('age_group')['chin_sagging'].value_counts().sort_index().values.sum())
chin_sagging.groupby('age_group')['chin_sagging'].value_counts().sort_index()
```

```python
# chin_sagging == 5인 것에서 나이대의 분포
chin_sagging[chin_sagging['chin_sagging'] == 5].groupby('age_group')['chin_sagging'].value_counts()
```

```python
# chin_sagging == 4인 것에서 나이대의 분포
chin_sagging[chin_sagging['chin_sagging'] == 4].groupby('age_group')['chin_sagging'].value_counts()
```

```python
# chin_sagging == 3인 것에서 나이대의 분포
chin_sagging[chin_sagging['chin_sagging'] == 3].groupby('age_group')['chin_sagging'].value_counts()
```

```python
# chin_sagging == 2인 것에서 나이대의 분포
chin_sagging[chin_sagging['chin_sagging'] == 2].groupby('age_group')['chin_sagging'].value_counts()
```

```python
# chin_sagging == 1인 것에서 나이대의 분포
chin_sagging[chin_sagging['chin_sagging'] == 1].groupby('age_group')['chin_sagging'].value_counts()
```

```python
# chin_sagging == 0인 것에서 나이대의 분포
chin_sagging[chin_sagging['chin_sagging'] == 0].groupby('age_group')['chin_sagging'].value_counts()
```

* 등급별 **나이 분포**를 확인해서 등급을 나눠준다.
        등급 = 0, 1 ==> 0
        등급 = 2, 3 ==> 1
        등급 = 4, 5, 6 ==> 2

```python
chin_sagging = chin_sagging.replace({'chin_sagging':1}, 0).replace({'chin_sagging':2}, 1).replace({'chin_sagging':3}, 1).replace({'chin_sagging':4}, 1).replace({'chin_sagging':5}, 1).replace({'chin_sagging':6}, 1)
chin_sagging.chin_sagging.value_counts()
```

```python
## 등급별로 나눠서 df 저장해서 train, val, test 나누기 ##
chin_sagging_0 = chin_sagging[chin_sagging.chin_sagging == 0]
chin_sagging_1 = chin_sagging[chin_sagging.chin_sagging == 1]
# chin_sagging_2 = chin_sagging[chin_sagging.chin_sagging == 2]
```

```python
def train_val_test_split(df):
    # 평가용 하나
    df_eval = df.groupby('age_group').sample(frac=0.2)
    df_train_val = df.drop(df_eval.index)
    # 훈련용 하나
    df_train = df_train_val.groupby('age_group').sample(frac=0.8)
    # 검증용 하나
    df_val = df_train_val.drop(df_train.index)
    return df_train, df_val, df_eval
```

```python
chin_sagging_0_train, chin_sagging_0_val, chin_sagging_0_eval = train_val_test_split(chin_sagging_0)
chin_sagging_1_train, chin_sagging_1_val, chin_sagging_1_eval = train_val_test_split(chin_sagging_1)
# chin_sagging_2_train, chin_sagging_2_val, chin_sagging_2_eval = train_val_test_split(chin_sagging_2)
```

```python
# 다시 훈련용, 검증용, 평가용 합치기
chin_sagging_train = pd.concat([chin_sagging_0_train, chin_sagging_1_train]).reset_index(drop=True)
chin_sagging_val = pd.concat([chin_sagging_0_val, chin_sagging_1_val]).reset_index(drop=True)
chin_sagging_eval = pd.concat([chin_sagging_0_eval, chin_sagging_1_eval]).reset_index(drop=True)
```

```python
# age_group 삭제
chin_sagging_train.drop(columns=['age_group'], inplace=True)
chin_sagging_val.drop(columns=['age_group'], inplace=True)
chin_sagging_eval.drop(columns=['age_group'], inplace=True)
```

```python
chin_sagging_train.to_csv(base_folder + '수린님/ResNet/classification/chin_sagging_train_class2.csv', index=False)
chin_sagging_val.to_csv(base_folder + '수린님/ResNet/classification/chin_sagging_val_class2.csv', index=False)
chin_sagging_eval.to_csv(base_folder + '수린님/ResNet/classification/chin_sagging_eval_class2.csv', index=False)
```
