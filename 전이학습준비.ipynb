## bbox 데이터 불러와서 csv로 만들기


```python
def make_bbox_data(source_folder, num):
    data = {}

    for file_path in glob.glob(os.path.join(source_folder, '*.txt')):
        image_filename = file_path.split('/')[-1].split('.')[0] + '.jpg'

        with open(file_path, 'r') as f:
            for line in f:
                parts = line.strip().split()

                if (len(parts) == 5) & (parts[0] == str(num)):
                    data[image_filename] = parts[1:]

    return data

def make_bbox_df(data):
    bbox_df = pd.DataFrame(data).T.reset_index()
    bbox_df.columns = ['image_filename', 'x_center', 'y_center', 'width', 'height']

    return bbox_df
```

## 측정 데이터와 합치기


```python
def merge_df(measurement_df, bbox_df):
    df = measurement_df.merge(bbox_df, on='image_filename')
    return df
```

```python
def change_type(df):

    # 데이터 타입 변경
    df = df.astype({
        'x_center': 'float',
        'y_center': 'float',
        'width': 'float',
        'height': 'float'
    })
    return df
```

```python
chin_train = change_type(merge_df(chin_measurement, chin_bbox_train))
chin_val = change_type(merge_df(chin_measurement, chin_bbox_val))
r_cheek_train = change_type(merge_df(r_cheek_measurement, r_cheek_bbox_train))
r_cheek_val = change_type(merge_df(r_cheek_measurement, r_cheek_bbox_val))
```
